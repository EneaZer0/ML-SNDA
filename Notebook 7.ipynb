In this final notebook, we are going to train a logistic classifier, and view its decision boundary in 3 dimensions.

We will use our custom iris dataset, but drop the last dimension. This is because we can't visualise data in more than 3 dimensions!

We now have 100 **3D** data points labelled as either **versicolor (0)** or **virginica (1)**. 40 of these points form the training set, 20 for validation, and 40 for test.

Formally, our training set is $\{\mathbf{{x}}_{i}\}_{i=1}^{N}$ ($N=40$) where each $\mathbf{x}_i\in \mathbb{R}^{3}$. Each datum has an associated label $y_i\in \{0, 1\}$ where $0$ corresponds to versicolor, and $1$ corresponds to virginica.

```python
import numpy as np # Import numpy

def load_iris():
    """
    This function loads in all our dataset splits.
    """
    D = np.load('iris_splits.npz')
    return D['X_train'], D['y_train'], D['X_val'], D['y_val'], D['X_test'], D['y_test']

X_train, y_train, X_val, y_val, X_test, y_test = load_iris()

# Drop last dimension
X_train = X_train[:,:-1]
X_val =  X_val[:,:-1]
X_test = X_test[:,:-1]
```

### Creating a logistic classifier

This will be very similar to the Linear classifier from the last notebook, but with the addition of a sigmoid function applied to the output.

```python
np.random.seed(5) # Fix the seed so we have the same random numbers each time we run this.

class LogisticClassifier:
    """
    This class creates a logistic classifer object with randomly initialised weight and bias.
    Its apply function can be used to apply the classifier to a data point
    """
    def __init__(self, num_dimensions=3):
        self.w = np.random.normal(size=num_dimensions) # Create a weight vector 
        self.b = np.random.normal(size=1) # Create a bias
        
        
    def sigmoid(self, z):
        """
        Sigmoid function defined for use in this classifier.
        """
        return 1/(1+np.exp(-z))
        
    def apply(self, x):
        """
        This function takes in a data point x, and applies the classifier to it.
        """
        return self.sigmoid(x@self.w.T + self.b) # Computes the classifier output 
```

Let's make a classifier object from this class, and apply it to a dummy data point. Remember that a logistic classifier takes a data point $\mathbf{x}_i$ and outputs

$$z_i = \sigma(\mathbf{x}_i\mathbf{w}^\top + b)$$ 

where $z_i\in \mathbb{R}^{1}$ and $\sigma()$ is the sigmoid function.

```python
classifier = LogisticClassifier() # Make classifier
x_i = np.array([1,2,3]) # dummy data point (we are now working in 3D)
z_i = classifier.apply(x_i)
print(z_i)
```

### Training a classifier

To train a classifier, we need 

- to devise a loss function, that is small when our classifier does what we want it to do 
- an optimisation algorithm to minimise that loss function

Following the video, we are going to use the **cross-entropy loss** averaged across our training data.

$$L = \frac{1}{N}\sum_{i=1}^{N}-y_i\log z_i - (1-y_i)\log (1-z_i)$$

We will optimise the parameters using gradient descent, so we need expressions for $\frac{\partial L}{\partial \mathbf{w}}$ and $\frac{\partial L}{\partial b}$.

In the tutorial sheet, you will derive this $\frac{\partial L}{\partial \mathbf{w}}$ in the 1D case. In the multi-dimensional case, these derivatives are given by:

$$\frac{\partial L}{\partial \mathbf{w}} = \frac{1}{N}\sum_{i=1}^{N} \mathbf{x}_i (z_i - y_i)$$

$$\frac{\partial L}{\partial b} = \frac{1}{N}\sum_{i=1}^{N} (z_i - y_i)$$

These expressions are very similar to those obtained for the gradients of a linear classifier with a MSE loss.

Let's now learn these parameters through gradient descent:

```python
def train_classifier_using_GD(classifier, X_train, y_train, alpha = 0.1, num_iters = 100):
    """
    This function takes in a logistic classifier object, training data, labels, and a learning
    rate alpha and performs iterations of gradient descent to update the classifier's
    parameters. It uses the gradients of the cross entropy loss. The function will change the
    parameter weights of the classifier object, and outputs the cross entropy loss for each iteration.
    Args:
        classifier : A logistic classifier object as defined above in this notebook.
        X_train : training set represented by a matrix (NxD) of N data points of dimensionality D
        y_train : labels for the training set, a vector N where each element i corresponds to the label of X_train[i]
        alpha : The learning rate, default 0.1
        num_iters : The number of iterations to perform gradient descent, default 100
    Example use:
        >>> loss_array = train_classifier_using_GD(classifier, X_train, y_train)
    """
    
    num_data_points, num_dimensions = X_train.shape # Get num data points, and dims
    loss_array = [] # Empty array to append the MSE loss at each iteration.
    
    for _ in range(num_iters):
        
        z = classifier.apply(X_train) # Apply the classifier to the training points
         
        dL_dw =  np.zeros(num_dimensions) # Create a zero array to populate
        dL_db = 0 # Zero value for bias
        
        for i in range(num_data_points):
            dL_dw += X_train[i]*(z[i]-y_train[i]) # Sum for each data point
            dL_db += (z[i]-y_train[i]) # Sum for each data point

        dL_dw *= 1/num_data_points # Multiply by 1/N
        dL_db *= 1/num_data_points # Multiply by 1/N
        
        # Now, update the parameters
        classifier.w -= alpha * dL_dw 
        classifier.b -= alpha * dL_db
        
        # (Re)compute loss and append to loss array
        z = classifier.apply(X_train) 
        loss = np.mean(-y_train * np.log(z) -(1-y_train)*np.log(1-z)) # Cross entropy loss expression
        loss_array.append(loss)

    return loss_array

loss_array = train_classifier_using_GD(classifier, X_train, y_train) # Perform gradient descent to learn our classifier!
```

We can plot the loss across iterations to see if the optimisation was successful.

```python
import matplotlib.pyplot as plt # Plotting code
plt.plot(loss_array)
plt.xlabel('Iteration')
plt.ylabel('Loss')
```

### Evaluating the classifier

Let's compute test accuracy, as we did in the last notebook.

```python
def compute_accuracy(outputs, true_labels):
    """
    Compares predicted labels in predictions to the actual labels in true_labels and computes an accuracy score.
    Args:
        predictions : an N-D array of predictions obtained from a classifier for some data points
        true_labels : an N-D array of ground-truth labels 
    Example use:
        >>> val_accuracy = compute_accuracy(val_predictions, y_val)
        >>> test_accuracy = compute_accuracy(test_predictions, y_test)
    """
    predictions = np.zeros_like(outputs)
    for i in range(len(outputs)):
        if outputs[i] > 0.5: # Classify as 1 if >0.5 and as 0 if <0.5
            predictions[i] = 1
        else:
            predictions[i] = 0

    return 100*np.sum(predictions==true_labels)/len(predictions)

outputs = classifier.apply(X_test)
acc = compute_accuracy(outputs, y_test)
print(f'Our trained logistic classifier is {acc:0.2f}% accurate.')
```

### Viewing the classifier's decision boundary

Let's start by visualising the test data in 3D space. 

```python
### Viewing the classifier's decision boundary

class0 = np.where(y_test==0) # Find test points in class 0
class1 = np.where(y_test==1) # Find test points in class 1
 
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_test[class0,0], X_test[class0,1], X_test[class0,2], color='b') # Plot class 0 test points in blue
ax.scatter(X_test[class1,0], X_test[class1,1], X_test[class1,2], color='r') # Plot class 1 test points in red

ax.set_xlabel('sepal length (standardised)') # Label your axes!
ax.set_ylabel('sepal width (standardised)') # Label your axes!
ax.set_zlabel('petal length (standardised)') # Label your axes!



plt.legend(['versicolor', 'virginica']) # Name classes in the legend
 
```

What exactly is our classifier doing? We know that for 2D data, the classifier draws a line that tries to separate data from the two classes. This is called its **decision boundary**.

In 3D the decision boundary of a classifier is a **plane**. We will now determine what this plane is, and draw it on the plot above.

Our classifier predicts that a point belongs to class 1 if its output is greater than 0.5, and that it belongs to class 0 if its output is less than 0.5. Its decision boundary is therefore defined by the plane

$$ 0.5 = \sigma(\mathbf{x}_i\mathbf{w}^\top + b)$$

The sigmoid function makes this quite tricky to work with, but we know how it behaves:

```python
# Let's plot the sigmoid function
k = np.linspace(-10,10,1000)
sig_k = 1/(1+np.exp(-k))
plt.plot(k,sig_k)
plt.xlabel('k')
plt.ylabel('sig(k)')
```

For inputs less than 0, the sigmoid function outputs values less than 0.5, and for inputs greater than 0, the sigmoid function outputs values greater than 0.5.

This means that we can rewrite our decision boundary as


$$ 0 = \mathbf{x}_i\mathbf{w}^\top + b$$

We can write the dot product as a sum in terms of vector elements to get 

$$ 0 = w_1 x_1 + w_2 x_2 + w_3 x_3+ b$$

We can rearrange this to solve for $x_3$

$$ x_3 = -\frac{w_1}{w_3} x_1 - \frac{w_2}{w_3} x_2 - \frac{b}{w_3}$$

We can now create a grid of $x_1$ and $x_2$ coordinates to produce their $x_3$ coordinates, and then plot this plane alongside our test data.

```python
x1, x2 = np.meshgrid(np.linspace(-2,2,10), np.linspace(-2,2,10)) # Generate a grid of x1 x2 points

w1, w2, w3, b = classifier.w[0], classifier.w[1], classifier.w[2], classifier.b # Get classifier params

x3 = -(w1/w3) * x1 - (w2/w3) * x2 - (b/w3) # Compute x3 for our x1 x2 grid

### Viewing the classifier's decision boundary
class0 = np.where(y_test==0) # Find test points in class 0
class1 = np.where(y_test==1) # Find test points in class 1
 
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(X_test[class0,0], X_test[class0,1], X_test[class0,2], color='b') # Plot class 0 test points in blue
ax.scatter(X_test[class1,0], X_test[class1,1], X_test[class1,2], color='r') # Plot class 1 test points in red

ax.set_xlabel('sepal length (standardised)') # Label your axes!
ax.set_ylabel('sepal width (standardised)') # Label your axes!
ax.set_zlabel('petal length (standardised)') # Label your axes!

plt.legend(['versicolor', 'virginica']) # Name classes in the legend

ax.plot_surface(x1,x2,x3, color='k',  alpha=0.7) # Draw our decision boundary plane

```

The plane largely succeeds at separating the points from each class in the test data. Neat!
